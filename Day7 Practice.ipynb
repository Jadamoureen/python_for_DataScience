```python
# r' means regular expressions. The last bracket is to turn everything into a hashtag. the .*? are there in case someone used them instead of the #

data['user_description'] = data.user_description.str.findall(r'#.*?(?=\s|$)') 
```

```python
data.dtypes
```

```python
def cats(df):

    text_cols = df.select_dtypes(include = ['object','bool'])
    
    for col in text_cols:
        
        df[col] = df[col].astype(str)
        
        df[col] = df[col].astype('category')
        
cats(data)
```

```python
def my_labeler(df):
    
    cols = ['truncated','favorited','retweeted','lang','user_screen_name','user_description','user_verified','month','dayofweek']
    
    from sklearn.preprocessing import LabelEncoder
    
    l_encoder = LabelEncoder()
    
    for each in cols:
        
        df[each] = l_encoder.fit_transform(df[each])
        
my_labeler(data)
```
